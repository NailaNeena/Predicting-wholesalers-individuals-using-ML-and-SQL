The code below is used for linear modeling to determine which variables 
impact the type of client the company has. We are looking to determine which 
clients should be called vs. emailed depending on if they are small businesses or not

```python
# importing libraries
import pandas as pd # data science essentials
import matplotlib.pyplot as plt
import seaborn as sns  # enhanced graphical output
import numpy as np # mathematical essentials
import statsmodels.formula.api as smf # regression modeling
import statsmodels.api as sm 
from sklearn.model_selection import train_test_split # train/test split
from sklearn.linear_model import LinearRegression

# setting pandas print options
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)


# specifying file name
file = './Retail Output_1.csv'
# reading the file into Python
flora = pd.read_csv(file)

flora.head(n = 10)

```

Completing an initial OLS regression model using the following variables: 

Y-Variable
total_price


X-Variable
invoice_count / product_count

```python
# building a base model

# INSTANTIATING a model type
lm_flora = smf.ols(formula = """  total_price ~ invoice_count""",
                                     data = flora)

# telling Python to FIT the data to the blueprint
lm_results = lm_flora.fit()


# printing a summary of the results
print(lm_results.summary())
```

```python
# building a base model

# INSTANTIATING a model type
lm_flora = smf.ols(formula = """  total_price ~ product_count""",
                                     data = flora)

# telling Python to FIT the data to the blueprint
lm_results = lm_flora.fit()


# printing a summary of the results
print(lm_results.summary())
```

**Initial Findings: 
Invoice count has a p-value higher than 0.05 and does not seem to have a 
strong correlation between total price. Whereas the number of products purchased 
does have a high adjusted r-squared (0.92) and p-value lower than 0.05.

```python
# defining a function for categorical boxplots
def categorical_boxplots(response, cat_var, data):
    """
	This function is designed to generate a boxplot for  can be used for categorical variables.
    Make sure matplotlib.pyplot and seaborn have been imported (as plt and sns).

    PARAMETERS
	----------
	response : str, response variable
	cat_var  : str, categorical variable
	data     : DataFrame of the response and categorical variables
	"""

    fig, ax = plt.subplots(figsize = (10, 8))
    
    sns.boxplot(x    = response,
                y    = cat_var,
                data = data)
    
    plt.suptitle("")
    plt.show()
```

```python
# one hot encoding categorical variables
one_hot_occupation           = pd.get_dummies(flora['occupation'])
one_hot_employment           = pd.get_dummies(flora['employment_type'])
one_hot_education            = pd.get_dummies(flora['education'])
one_hot_marital_status       = pd.get_dummies(flora['marital_status'])
one_hot_relationship         = pd.get_dummies(flora['relationship_in_household'])
one_hot_race                 = pd.get_dummies(flora['race'])
one_hot_country              = pd.get_dummies(flora['original_country'])
one_hot_client               = pd.get_dummies(flora['type_of_client'])

# dropping categorical variables after they've been encoded
flora = flora.drop('occupation', axis = 1)
flora = flora.drop('employment_type', axis = 1)
flora = flora.drop('education', axis = 1)
flora = flora.drop('marital_status', axis = 1)
flora = flora.drop('relationship_in_household', axis = 1)
flora = flora.drop('race', axis = 1)
flora = flora.drop('original_country', axis = 1)
flora = flora.drop('type_of_client', axis = 1)

# joining codings together
flora = flora.join([one_hot_occupation, one_hot_employment, one_hot_education, one_hot_marital_status, one_hot_relationship, one_hot_race, one_hot_country, one_hot_client])
```

```python
# preparing explanatory variable data
flora_data   = flora.drop(['total_price',
                               'log_total_price',
                               'product_count',
                               'log_product_count'],
                                axis = 1)


# preparing response variable data
flora_target = flora.loc[ : , 'total_price']
log_flora_target = flora.loc[ : , 'log_total_price'] # ready for use later


# preparing training and testing sets (all letters are lowercase)
x_train, x_test, y_train, y_test = train_test_split(
            flora_data,
            flora_target,
            test_size = 0.25,
            random_state = 219)


# checking the shapes of the datasets
print(f"""
Training Data
-------------
X-side: {x_train.shape}
y-side: {y_train.shape[0]}


Testing Data
------------
X-side: {x_test.shape}
y-side: {y_test.shape[0]}
""")
```

```python
# building a full model

# blueprinting a model type
lm_full = smf.ols(formula = """ total_price ~
                               customer_id +
                                sex_at_birth +
                                completed_years_of_education +
                                invoice_count +
                                date +
                                log_product_count +
                                m_sex_at_birth +
                                m_occupation +
                                m_employment_type +
                                m_education +
                                m_marital_status +
                                m_relationship_in_household +
                                m_race +
                                m_original_country +
                                m_type_of_client""",
                                data = flora)


# telling Python to run the data through the blueprint
results_full = lm_full.fit()


# printing the results
results_full.summary()
```

```python
# applying model in scikit-learn

# Preparing a DataFrame based the the analysis above
ols_data   = flora.loc[ : , x_variables]


# Preparing the target variable
flora_target = flora.loc[ : , 'total_price']


###############################################
## setting up more than one train-test split ##
###############################################
# FULL X-dataset (normal Y)
x_train_FULL, x_test_FULL, y_train_FULL, y_test_FULL = train_test_split(
            flora_data,     # x-variables
            flora_target,   # y-variable
            test_size = 0.25,
            random_state = 219)


# OLS p-value x-dataset (normal Y)
x_train_OLS, x_test_OLS, y_train_OLS, y_test_OLS = train_test_split(
            ols_data,         # x-variables
            flora_target,   # y-variable
            test_size = 0.25,
            random_state = 219)
```

Linear regression not running because the dummy variables have yet to be converted.

```python
# INSTANTIATING a model object
lr = LinearRegression()


# FITTING to the training data
lr_fit = lr.fit(x_train_OLS, y_train_OLS)


# PREDICTING on new data
lr_pred = lr_fit.predict(x_test_OLS)


# SCORING the results
print('OLS Training Score :', lr.score(x_train_OLS, y_train_OLS).round(4))
print('OLS Testing Score  :', lr.score(x_test_OLS, y_test_OLS).round(4))


# saving scoring data for future use
lr_train_score = lr.score(x_train_OLS, y_train_OLS).round(4) # using R-square
lr_test_score  = lr.score(x_test_OLS, y_test_OLS).round(4)   # using R-square


# displaying and saving the gap between training and testing
print('OLS Train-Test Gap :', abs(lr_train_score - lr_test_score).round(4))
lr_test_gap = abs(lr_train_score - lr_test_score).round(4)
```

```python
# zipping each feature name to its coefficient
lr_model_values = zip(flora_data[x_variables].columns,
                      lr_fit.coef_.round(decimals = 2))


# setting up a placeholder list to store model features
lr_model_lst = [('intercept', lr_fit.intercept_.round(decimals = 2))]


# printing out each feature-coefficient pair one by one
for val in lr_model_values:
    lr_model_lst.append(val)
    

# checking the results
for pair in lr_model_lst:
    print(pair)
```
